15. List - Pair Sum Problem
Given a list of numbers and a target sum, find all pairs in the list that add up to the target.
Example:
Input: [2, 4, 3, 7, 1, 9], Target: 10
Output: [(3, 7), (1, 9)]

```python
l=[2, 4, 3, 7, 1, 9]
tar =10
ans=[]
seen=set()

for n in l:
  second =tar-n
  if second in seen:
      ans.append((n,second))
  seen.add(n)
print(ans)
```
